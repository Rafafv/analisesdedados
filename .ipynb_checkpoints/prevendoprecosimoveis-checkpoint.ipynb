 # Explorando Dataset Boston Housing
 
**** Dataset https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_boston.html

```python
#importando os modulos necessarios
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
%matplotlib inline
```

```python
# o dataset boston ja esta disponivel no scikit-learn, precisa apenas carrega-lo
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
#verificando o tipo da variavel boston
type(boston)
```

```python
#visualizando o shape do dataset, neste caso 506 linhas e 13 colunas
boston.data.shape
```

```python
#descricao do dataset
print(boston.DESCR)
```

```python
print(boston.feature_names)
```

```python
#convertendo o dataset em um DataFrame pandas
df = pd.DataFrame(boston.data)
df.head()
```

```python
#convertendo o titulo das colunas
df.columns = boston.feature_names
df.head()
```

```python
boston.target
```

```python
df['PRICE'] = boston.target
df.head()
```

# Prevendo o Preço das casas em Boston

 Y - variável dependente (preço das casas em Boston)
 X - variáveis independentes ou exploratórias (todas as outras características da casa)

```python
#importando o módulo de regressao linear
from sklearn.linear_model import LinearRegression
```

```python
X = df.drop('PRICE', axis = 1)
```

```python
Y = df.PRICE
```

```python
plt.scatter(df.RM, Y)
plt.xlabel("Média do Número de Quartos por Casa")
plt.ylabel("Preço da Casa")
plt.title("Relação entre Número de Quartos e Preço")
plt.show()
```

```python
# Criando objeto de regressao linear
regr = LinearRegression()
```

```python
#tipo do objeto
type(regr)
```

```python
#treinando o modelo
regr.fit(X, Y)
```

```python
#coeficientes
print("Coeficientes: ", regr.intercept_)
print("Número de Coeficientes: ", len(regr.coef_))
```

```python
#prevendo o preço da casa
regr.predict(X)
```

```python
#comparando preços originais x preços previstos
plt.scatter(df.PRICE, regr.predict(X))
plt.xlabel("Preoço Original")
plt.ylabel("Preço Previsto")
plt.title("Preço Original x Preço Previsto")
plt.show()
```

Existem alguns erros na predição do preço das casas

```python
#calcular o MSE(Mean Squared Error)
mse1 = np.mean((df.PRICE - regr.predict(X))** 2)
print(mse1)
```

```python
#aplicando regressao linear para apenas uma variavel e calculando o MSE
regr = LinearRegression()
regr.fit(X[['PTRATIO']], df.PRICE)
mse2 = np.mean((df.PRICE - regr.predict(X[['PTRATIO']])) ** 2)
print(mse2)
```

# O MSE aumentou, indicando que uma única caracteristica não é um bom predictor para o preço das casas.

```python
#dividindo X em dados de treino e de teste
X_treino = X[:-50]
X_teste = X[-50:]

#dividindo Y em dados de treino e de teste
Y_treino = df.PRICE[:-50]
Y_teste = df.PRICE[-50:]

#imprimindo o shape dos datasets
print(X_treino.shape, X_teste.shape, Y_treino.shape, Y_teste.shape)
```

# A forma correta seria usando o scikit-learn para a divisão dos valores

```python
from sklearn.model_selection import train_test_split
```

```python
#dividindo X e Y em dados de treino e de teste
X_treino, X_teste, Y_treino, Y_teste = train_test_split(X, df.PRICE, test_size = 0.30, random_state =5)
```

```python
#imprimindo o shape dos dataset
print(X_treino.shape, X_teste.shape, Y_treino.shape, Y_teste.shape)
```

```python
#construindo um modelo de regressao
regr = LinearRegression()
```

```python
#treinando o modelo
regr.fit(X_treino, Y_treino)
```

```python
#definindo os dados de treino e teste
pred_treino = regr.predict(X_treino)
pred_teste = regr.predict(X_teste)
```

```python
#comparando preços originais x preços previstos
plt.scatter(regr.predict(X_treino), regr.predict(X_treino) - Y_treino, c = 'b', s = 40, alpha = 0.5)
plt.scatter(regr.predict(X_teste), regr.predict(X_teste) - Y_teste, c = 'g', s = 40, alpha = 0.5)
plt.hlines(y = 0, xmin = 0, xmax = 50)
plt.ylabel("Resíduos")
plt.title("Residual Plot - Treino(Azul), Teste(Verde)")
plt.show()
```
